```python
class Solution(object):
    def isMatch(self, s, p):
        """
        :type s: str
        :type p: str
        :rtype: bool
        """

        if ('*' not in p) and ('.' not in p):
            if s != p:
                print('Im here')
                return False
            else:
                return True
        
        word = s
        path = self.generate_graph(p)

        num_conditions = len(path)
        print('Num conditions:', num_conditions)

        flag = 0
        iter = 0

        for i in range(0, len(word) - 1):

            # if (word[i] == path[iter][0] or '.' in path[iter][0]) and (word[i+1] in path[iter][1] or '.' in path[iter][1]) and word[i] not in path[iter][1]:
            if (word[i] == path[iter][0] or '.' in path[iter][0]) and (word[i+1] in path[iter][1] or '.' in path[iter][1]) and word[i] != word[i+1]:
                iter += 1
            elif word[i] == word[i+1] and (word[i] == path[iter][0] or '.' in path[iter][0]) and (word[i] in path[iter][1] or '.' in path[iter][1]) and iter == num_conditions - 1:
                iter += 1
            elif (word[i] == path[iter][0] or '.' in path[iter][0]) and word[i+1] == word[i] and (word[i] in path[iter][1] or '.' in path[iter][1]):
                iter += 0
            else:
                iter = 0

            print('iter:', iter, '| i:', i, '| word:', word[:i+1], '| i:', word[i], '| i+1:', word[i+1])

            if iter == num_conditions:
                flag = 1
                iter = 0

        if flag == 1:
            print('True')
            return True
        else:
            print('False')
            return False

    def generate_graph(self, pattern):

        pattern_size = len(pattern)

        if pattern_size == 1:
            return [[pattern[0]]]

        graph_structure = []

        for i in range(0, pattern_size):

            try:
                #if pattern[i] == '*':
                    # graph_structure.append([pattern[i-1], pattern[i-1]])
                #    continue

                if (pattern[i].isalpha() or pattern[i] == '.') and (pattern[i+1].isalpha() or pattern[i+1] == '.'):
                    graph_structure.append([pattern[i], pattern[i+1]])

                elif (pattern[i].isalpha() or pattern[i] == '.') and pattern[i+1] == '*':
                    try:
                        graph_structure.append([pattern[i], f'{pattern[i+2]}, {pattern[i]}'])
                    except IndexError:
                        graph_structure.append([pattern[i], pattern[i]])

                else:
                    continue

            except IndexError:
                continue
            
        print('Graph structure:', graph_structure)
        return graph_structure
```
