```python
# FIRST SOLUTION

class Solution:
    def threeSum(self, nums: list[int]) -> dict[int, str]:
        
        if nums == []:
            return []
        
        combinations_list = []

        nums.sort()

        iter = 0

        while iter != len(nums) - 1:
            
            l_pointer = iter + 1
            r_pointer = len(nums) - 1

            while l_pointer != r_pointer:

                if nums[iter] + nums[l_pointer] + nums[r_pointer] < 0:
                    l_pointer += 1
                elif nums[iter] + nums[l_pointer] + nums[r_pointer] > 0:
                    r_pointer -= 1
                else:
                    combinations_list.append([nums[iter], nums[l_pointer], nums[r_pointer]])
                    if l_pointer + 1 != r_pointer:
                        l_pointer += 1
                        r_pointer -= 1
                    else:
                        break

            iter += 1

        final_list = []

        for list in combinations_list:
            if list not in final_list:
                final_list.append(list)

        return final_list
```

```python
# SECOND SOLUTION

class Solution:
    def threeSum(self, nums: list[int]) -> dict[int, str]:
        
        if nums == []:
            return []
        
        combinations_list = []

        nums.sort()

        iter = 0

        while iter != len(nums) - 1:
            
            if iter > 0 and nums[iter] == nums[iter-1]:
                iter += 1
                continue

            l_pointer = iter + 1
            r_pointer = len(nums) - 1

            while l_pointer != r_pointer:

                if nums[iter] + nums[l_pointer] + nums[r_pointer] < 0:
                    l_pointer += 1

                elif nums[iter] + nums[l_pointer] + nums[r_pointer] > 0:
                    r_pointer -= 1

                else:
                    combinations_list.append([nums[iter], nums[l_pointer], nums[r_pointer]])
                    l_pointer += 1

                    while nums[l_pointer] == nums[l_pointer - 1] and l_pointer < r_pointer:
                        l_pointer += 1

            iter += 1

        return combinations_list
```
