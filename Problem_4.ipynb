```python
# FIRST SOLUTION: O(m+n)

class Solution(object):
    def findMedianSortedArrays(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: float
        """

        sorted_array = []
        
        nums1_pos = 0
        nums2_pos = 0

        if len(nums1) == 0 and len(nums2) == 0:
            return 0
        
        elif len(nums1) == 0 or len(nums2)== 0:
            sorted_array = nums1 + nums2
            
        else:
            while True:
                if nums1[nums1_pos] >= nums2[nums2_pos]:
                    sorted_array.append(nums2[nums2_pos])
                    nums2_pos += 1
                else:
                    sorted_array.append(nums1[nums1_pos])
                    nums1_pos += 1

                if nums1_pos == len(nums1):
                    sorted_array += nums2[nums2_pos:]
                    break

                if nums2_pos == len(nums2):
                    sorted_array += nums1[nums1_pos:]
                    break

        if len(sorted_array) % 2 == 0:
            return (float(sorted_array[len(sorted_array) // 2]) + sorted_array[len(sorted_array) // 2 - 1]) / 2
        else:
            return sorted_array[len(sorted_array) // 2]
        
```

```python
# SECOND SOLUTION: O(m+n)

class Solution(object):
    def findMedianSortedArrays(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: float
        """

        size = len(nums1 + nums2)

        if size % 2 == 0:
            type = True
        else: 
            type = False 

        size_1 = len(nums1)
        pointer_1 = 0
        size_2 = len(nums2)
        pointer_2 = 0

        value_front = 0
        
        while pointer_1 + pointer_2 != (size // 2) + 1:
            
            value_back = value_front

            if pointer_2 < size_2 and pointer_1 < size_1:
                if nums1[pointer_1] >= nums2[pointer_2]:
                    value_front = nums2[pointer_2]
                    pointer_2 += 1
                else:
                    value_front = nums1[pointer_1]
                    pointer_1 += 1

            elif pointer_2 < size_2:
                value_front = nums2[pointer_2]
                pointer_2 += 1
            
            else:
                value_front = nums1[pointer_1]
                pointer_1 += 1
        
        if type:
            return (value_front + value_back) / 2.0
        else:
            return value_front
```
