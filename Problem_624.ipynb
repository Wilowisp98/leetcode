```python
# FIRST SOLUTION: O(n^2)

class Solution(object):
    def maxDistance(self, arrays):
        """
        :type arrays: List[List[int]]
        :rtype: int
        """
        size = len(arrays)

        min_list = []
        max_list = []

        for list in arrays:
            min_list.append(list[0])
            max_list.append(list[-1])
        
        sliding_window = []
        for i in range(1,size):
            sliding_window.append(min_list[i:] + min_list[0:i])

        results = []

        for window in sliding_window:
            for i in range(0,size):
                results.append(abs(max_list[i] - window[i]))

        return max(results)
```

```python
# SECOND SOLUTION: O(n)

class Solution(object):
    def maxDistance(self, arrays):
        """
        :type arrays: List[List[int]]
        :rtype: int
        """

        min_list = []
        max_list = []

        for list in arrays:
            min_list.append(list[0])
            max_list.append(list[-1])

        maxSet = self.getMaxs(max_list)
        minSet = self.getMins(min_list)
        print(maxSet)
        print(minSet)
        if maxSet['max'][1] != minSet['min'][1]:
            return abs(maxSet['max'][0] - minSet['min'][0])
        
        else:
            return max(abs(maxSet['max'][0] - minSet['second_min'][0]), abs(maxSet['second_max'][0] - minSet['min'][0]))

    
    def getMins(self, min_list):
        if min_list[0] < min_list[1]:
            min_set = {'min':[min_list[0],0],'second_min':[min_list[1],1]}
        else:
            min_set = {'min':[min_list[1],1],'second_min':[min_list[0],0]}
        for i in range(2,len(min_list)):
            if min_list[i] < min_set['min'][0]:
                min_set['second_min'] = min_set['min']
                min_set['min'] = [min_list[i],i]
            elif min_list[i] < min_set['second_min'][0]:
                min_set['second_min'] = [min_list[i],i]
            
        return min_set
    
    def getMaxs(self, max_list):
        if max_list[0] > max_list[1]:
            max_set = {'max':[max_list[0],0],'second_max':[max_list[1],1]}
        else:
            max_set = {'max':[max_list[1],1],'second_max':[max_list[0],0]}
        for i in range(2,len(max_list)):
            if max_list[i] > max_set['max'][0]:
                max_set['second_max'] = max_set['max']
                max_set['max'] = [max_list[i],i]
            elif max_list[i] > max_set['second_max'][0]:
                max_set['second_max'] = [max_list[i],i]
            
        return max_set
```

```python
# THIRD SOLUTION (from leetcode)

class Solution(object):
    def maxDistance(self, arrays):
        smallest = arrays[0][0]
        biggest = arrays[0][-1]
        max_distance = 0

        for i in range(1, len(arrays)):
            arr = arrays[i]
            max_distance = max(max_distance, abs(arr[-1] - smallest), abs(biggest - arr[0]))
            smallest = min(smallest, arr[0])
            biggest = max(biggest, arr[-1])

        return max_distance
```
